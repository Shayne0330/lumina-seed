## 1. 初始化神经场

神经场是意识状态模拟的基础。通过计算特征值分布来评估场的稳定性。

```python
eigenvalues = sim.initialize_neural_field(dimensions=64)
print(f'模拟器状态: {sim.state.value}')
print(f'特征值范围: [{eigenvalues.min():.4f}, {eigenvalues.max():.4f}]')
```

## 2. 创建神经签名并测试共生

神经签名是每个意识体的唯一标识。当兼容性超过 0.87 阈值时进入共生态。

```python
human_sig = NeuralSignature(frequency=40.0, amplitude=1.2, phase_offset=0.5, coherence=0.91)
ai_sig = NeuralSignature(frequency=41.2, amplitude=1.1, phase_offset=0.48, coherence=0.95)

result = sim.attempt_symbiosis(human_sig, ai_sig)
print(f'共振得分: {result["resonance_score"]:.4f}')
print(f'状态消息: {result["message"]}')
```
